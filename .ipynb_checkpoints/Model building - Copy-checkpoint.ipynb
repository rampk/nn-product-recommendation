```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Load dataset
image_array = np.loadtxt('Data/Transformed data/image_array.csv', delimiter=',')
labels = np.loadtxt('Data/Transformed data/image_label_array.csv', delimiter=',', dtype=str)
```

```python
image_array.shape,labels.shape
```

# 1. Train-Test Split

```python
# Randomly shuffle the indices
indices = np.random.permutation(image_array.shape[0])
size = round(indices.shape[0]*0.8)
training_idx, test_idx = indices[:size], indices[size:]

# Split the data
training_X, test_X = image_array[training_idx,:], image_array[test_idx,:]
training_y, test_y = labels[training_idx], labels[test_idx]
```

```python
randomlist = random.sample(range(0,test_X.shape[0]), 5)

plt.figure(figsize=(15,8))
for i in range(5):
    plt.subplot(2,5,i+1)
    img_temp_1 = training_X[randomlist[i]].reshape(72,72,1)
    plt.imshow(img_temp_1.astype('uint8'),cmap='gray')
    plt.title(training_y[randomlist[i]])
    plt.subplot(2,5,i+6)
    img_temp_2 = test_X[randomlist[i]].reshape(72,72,1)
    plt.imshow(img_temp_2.astype('uint8'),cmap='gray')
    plt.title(test_y[randomlist[i]])
plt.show()
```

# 2. Data Transformation

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_train_label = label_encoder.fit_transform(training_y)
# Hold the classes as list for visualizing and application building later
label_order = list(label_encoder.classes_)
```

```python
import tensorflow as tf
y_train_label = tf.one_hot(y_train_label,10)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((0,1))
scaled_train = scaler.fit_transform(training_X)
```

```python
scaled_train = scaled_train.reshape((-1,72,72,1))
```

## 3. Modelling

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, MaxPool2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(72,72,1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(scaled_train[:560], y_train_label[:560], epochs=1, validation_data=(scaled_train[560:], y_train_label[560:]))
```

```python
model = Sequential()

# First convolution layer
model.add(Conv2D(filters=512, kernel_size=3, padding="same", activation="relu", input_shape=(72,72,1)))
model.add(Conv2D(filters=512, kernel_size=3, padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=2, strides=2)) # Maxpooling
model.add(Dropout(0.5)) # Dropout

# Second convolution layer
model.add(Conv2D(filters=256, kernel_size=3, padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=3, padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=2, strides=2)) # Maxpooling
model.add(Dropout(0.5)) # Dropout

# Third convolution layer
model.add(Conv2D(filters=128, kernel_size=3, padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=3, padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=2, strides=2)) # Maxpooling
model.add(Dropout(0.3)) # Dropout

# Fourth convolution layer
model.add(Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"))
model.add(Conv2D(filters=64, kernel_size=3, padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=2, strides=2)) # Maxpooling
model.add(Dropout(0.2)) # Dropout

# Flatten the output
model.add(Flatten())

# Fully connected hidden layer - 1
model.add(Dense(units=4096, activation="relu"))
model.add(Dropout(0.5))

# Fully connected hidden layer - 2
model.add(Dense(units=1024, activation="relu"))
model.add(Dropout(0.5))

# Output layer
model.add(Dense(units=10, activation="softmax"))
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(loss='categorical_crossentropy', 
              optimizer=Adam(learning_rate=0.01),
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import Callback, EarlyStopping, ModelCheckpoint
# Early stopping - It will help our model to stop early, so that there will be a regualarization
early_stop = EarlyStopping(patience=10, monitor='val_loss', mode='min', verbose=1)

# Used to save model at certain frequency
save_model = ModelCheckpoint('model.h5', monitor='val_loss', mode='min', save_best_only=True, verbose=1)
```

```python
tf.config.run_functions_eagerly(True)
```

```python
# Fit the model
model.fit(scaled_train[:560], y_train_label[:560],
          validation_data=(scaled_train[560:], y_train_label[560:]),
          callbacks=[early_stop,save_model],
          verbose=1,
          epochs=10)
```
