```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Load dataset
image_array = np.loadtxt('Data/Transformed data/image_array.csv', delimiter=',')
labels = np.loadtxt('Data/Transformed data/image_label_array.csv', delimiter=',', dtype=str)
```

```python
image_array.shape,labels.shape
```

# 1. Train-Test Split

```python
# Randomly shuffle the indices
indices = np.random.permutation(image_array.shape[0])
size = round(indices.shape[0]*0.8)
training_idx, test_idx = indices[:size], indices[size:]

# Split the data
training_X, test_X = image_array[training_idx,:], image_array[test_idx,:]
training_y, test_y = labels[training_idx], labels[test_idx]
```

```python
randomlist = random.sample(range(0,test_X.shape[0]), 5)

plt.figure(figsize=(15,8))
for i in range(5):
    plt.subplot(2,5,i+1)
    img_temp_1 = training_X[randomlist[i]].reshape(72,72,1)
    plt.imshow(img_temp_1.astype('uint8'),cmap='gray')
    plt.title(training_y[randomlist[i]])
    plt.subplot(2,5,i+6)
    img_temp_2 = test_X[randomlist[i]].reshape(72,72,1)
    plt.imshow(img_temp_2.astype('uint8'),cmap='gray')
    plt.title(test_y[randomlist[i]])
plt.show()
```

# 2. Data Transformation

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
training_y = label_encoder.fit_transform(training_y)
# Hold the classes as list for visualizing and application building later
label_order = list(label_encoder.classes_)
```

```python
import tensorflow as tf
# Converting into one-hot vectors
y_train_label = tf.one_hot(training_y,10)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((0,1))
scaled_train = scaler.fit_transform(training_X)
```

```python
scaled_train = scaled_train.reshape((-1,72,72,1))
```

## 3. Modelling

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, MaxPool2D,ZeroPadding2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
from tensorflow.keras.optimizers import SGD
```

### VGG16 model

```python
vgg16_model = Sequential()
vgg16_model.add(Conv2D(input_shape=(72,72,1),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
vgg16_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
vgg16_model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
vgg16_model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
vgg16_model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
vgg16_model.add(Flatten())
vgg16_model.add(Dense(units=4096,activation="relu"))
vgg16_model.add(Dense(units=4096,activation="relu"))
vgg16_model.add(Dense(units=1024,activation="relu"))
vgg16_model.add(Dropout(0.5))
vgg16_model.add(Dense(units=10, activation="softmax"))
```

```python
vgg16_model.summary()
```

```python
# Training the model
opt = SGD(lr=0.1, decay=0.1 / 10)
vgg16_model.compile(loss='mean_squared_error',
              optimizer = opt,metrics=["accuracy"])
```

```python
train_set_size = int(scaled_train.shape[0] * .9)
history_1 = vgg16_model.fit(scaled_train[:train_set_size], y_train_label[:train_set_size],
          validation_data=(scaled_train[train_set_size:], y_train_label[train_set_size:]),epochs=100)
```

### VGG19

```python
vgg19 = Sequential()
vgg19.add(ZeroPadding2D((1,1),input_shape=(72,72,1)))
vgg19.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
vgg19.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
vgg19.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=256, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=256, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=256, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=256, kernel_size=(3,3), activation='relu'))
vgg19.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(ZeroPadding2D((1,1)))
vgg19.add(Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
vgg19.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

vgg19.add(Flatten())
vgg19.add(Dense(4096, activation='relu'))
vgg19.add(Dropout(0.5))
vgg19.add(Dense(4096, activation='relu'))
vgg19.add(Dropout(0.5))
vgg19.add(Dense(units=10, activation="softmax"))
```

```python
vgg19.summary()
```

```python
from tensorflow.keras.optimizers import Adam

opt = Adam(lr=0.1, decay=0.01)
vgg19.compile(loss='mean_squared_error',
              optimizer=opt,metrics=["accuracy"])
```

```python
history_2 = vgg19.fit(scaled_train[:train_set_size], y_train_label[:train_set_size],
          validation_data=(scaled_train[train_set_size:], y_train_label[train_set_size:]),epochs=100)
```

### Custom model 1

```python
model_custom_1 = Sequential()
model_custom_1.add(Conv2D(32, (3, 3), padding="same",activation="relu",input_shape=(72,72,1), strides=(1,1)))
model_custom_1.add(MaxPooling2D(pool_size=(3, 3)))

model_custom_1.add(Conv2D(64, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(Conv2D(64, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(MaxPooling2D(pool_size=(2, 2)))

model_custom_1.add(Conv2D(128, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(Conv2D(128, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(MaxPooling2D(pool_size=(2, 2)))

model_custom_1.add(Conv2D(512, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(Conv2D(512, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(Conv2D(512, (3, 3), padding="same",activation="relu", strides=(1,1)))
model_custom_1.add(MaxPooling2D(pool_size=(2, 2)))

model_custom_1.add(Flatten())
model_custom_1.add(Dense(1024, activation="relu"))
model_custom_1.add(Dropout(0.5))
model_custom_1.add(Dense(10,activation="softmax" ))
```

```python
model_custom_1.summary()
```

```python
opt = SGD(lr=0.1)
model_custom_1.compile(loss='mean_squared_error',
              optimizer = opt,metrics=["accuracy"])

history_3 = model_custom_1.fit(scaled_train[:train_set_size], y_train_label[:train_set_size],
          validation_data=(scaled_train[train_set_size:], y_train_label[train_set_size:]),epochs=100)
```

### Custom model 2

```python
model_custom2 = Sequential()
model_custom2.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(72,72,1)))
model_custom2.add(Conv2D(64, (3, 3), activation='relu'))
model_custom2.add(MaxPooling2D(pool_size=(2, 2)))
model_custom2.add(Dropout(0.25))
model_custom2.add(Conv2D(64, (3, 3), activation='relu'))
model_custom2.add(MaxPooling2D(pool_size=(2, 2)))
model_custom2.add(Dropout(0.25))
model_custom2.add(Flatten())
model_custom2.add(Dense(128, activation='relu'))
model_custom2.add(Dropout(0.5))
model_custom2.add(Dense(10, activation='softmax'))
```

```python
model_custom2.summary()
```

```python
model_custom2.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
history_4 = model_custom2.fit(scaled_train[:train_set_size], y_train_label[:train_set_size],
          validation_data=(scaled_train[train_set_size:], y_train_label[train_set_size:]),epochs=100)
```

### Test set

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
# Reshaping test set
test = scaler.transform(test_X)
test = test.reshape(-1,72,72,1)
true_test = label_encoder.transform(test_y)
```

```python
accuracy_test = []
plt.figure(figsize=(15,15))
for i, model in enumerate(models):
    plt.subplot(2,2,i+1)
    y_predict = models[i].predict_classes(test)
    cmatrix = confusion_matrix(true_test,y_predict)
    accuracy_test.append(accuracy_score(true_test,y_predict))
    
    # create heatmap
    sns.heatmap(pd.DataFrame(cmatrix,columns=label_order,index=label_order), annot=True, fmt='g')
    plt.title(names[i])
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
```
